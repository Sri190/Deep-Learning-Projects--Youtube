```python
#Import All The Libraries
import pandas as pd
import numpy as np
import tensorflow as tf
tf.random.set_seed(3)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow import keras

#Import The Dataset
df=pd.read_csv("ML13.csv")

#Data Cleaning
def mode_value(df,col):
    m=df[col].mode()[0]
    df.fillna({col:m},inplace=True)
def mean_value(df,col):
    m=df[col].mean()
    df.fillna({col:m},inplace=True)
num_col=[col for col in df.columns if df[col].dtype=="float64"]
obj_col=[col for col in df.columns if df[col].dtype=="object"]
for col in num_col:
    mean_value(df,col)
for col in obj_col:
    mode_value(df,col)
df['Sex']=df['Sex'].map({'male':1,'female':0})
df['Embarked']=df['Embarked'].map({'S':0,'C':1,'Q':2})

#Data Preprocessing
scaler=StandardScaler()
X=df.drop({'PassengerId','Ticket','Cabin','Name','Survived'},axis=1)
y=df['Survived']
x=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)

#Train-Test Splitting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)

#Model Training.
model = keras.Sequential([
    keras.layers.Input(shape=(7,)),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
early_stop = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)

history = model.fit(
    x_train, y_train,
    validation_split=0.1,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, reduce_lr]
)

#Model Accuracy
loss,Accuracy=model.evaluate(x_test,y_test)
print(f"Test Loss :{loss:.4f}")
print(f"Test Accuracy :{Accuracy:.4f}")
```
